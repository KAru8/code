2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#ваш код здесь

# С помощью lambda функции забираем первый символ строки
# и создаем таким образом столбец "Пол"

resume_base['Пол'] = resume_base['Пол, возраст'].apply(lambda x: x[0])


def to_split(string, delim=' '):
    """Функция для разделения строки по заданному символу

    Args:
        string (str): Строка, обрабатываемая функцией. 
        delim (str, optional): символ разделителя. По умолчанию - пробел

    Returns:
        massive: Список элементов - строк, полученных после разделения
    """
    massive = string.split(delim)
    return massive

# Применяем функцию to_split к столбцу 'Пол, возраст'

resume_base['Пол, возраст'] = resume_base['Пол, возраст'].apply(lambda x:
    to_split(x, delim=','))

# Далее из списка выбираем элемент с индексом 1, содержащий
# текстовую информацию о возрасте

resume_base['Возраст'] = resume_base['Пол, возраст'].apply(lambda x: x[1])


def to_integer(string):
    """Функция для выборки цифр из текста

    Args:
        string (str): Строка, обрабатываемая функцией. 
        
    Returns:
        int: Челое число
    """
    numb = []
    for symb in string:
        if symb in '0123456789':
            numb.append(symb)
    return int(''.join(numb))

# Применяем функцию to_integer к столбцу 'Возраст'
# и после удаляем исходный столбец 'Пол, возраст'

resume_base['Возраст'] = resume_base['Возраст'].apply(to_integer) 
resume_base = resume_base.drop('Пол, возраст', axis=1) 

display(f"Средний возраст соискателей:{resume_base['Возраст'].mean():.0f} лет")
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

def get_experience(arg):
    
    """Функция для выражения опыта работы в количестве месяцев

    Args:
        arg: обрабатываемая фукнцией переменная  
        

    Returns:
        NaN -  при выполнении условия arg == "Не указано" или np.nan
        month_exp - Целое число месяцев (int)
    """
       
    if arg == 'Не указано' or arg is np.nan:
        return np.nan
            
    else:
        arg = str(arg) 
        arg_exp = arg[:22] 
        year, month = '', ''
        year_list, month_list = [], []
    
        if 'год' in arg_exp:
            year, month = arg_exp.split('год')
        elif 'лет' in arg_exp:
            year, month = arg_exp.split('лет')
        else:
            month = arg_exp
    
        for symb in year:
            if symb in '0123456789':
                year_list.append(int(symb))
    
        for symb in month:
            if symb in '0123456789':
                month_list.append(int(symb))  
    
        if len(year_list) > 1:
            year_int = (year_list[0] * 10 * 12) + (year_list[1] * 12)
        elif len(year_list) == 1:
            year_int = year_list[0] * 12
        else:
            year_int = 0
        
        if len(month_list) > 1:
            month_int = (month_list[0] * 10) + (month_list[1])
        elif len(month_list) == 1:
            month_int = month_list[0]         
        else:
            month_int = 0           
        
        month_exp = year_int  + month_int
          
        return month_exp
    
    
# Применение функции get_experience к столбцу 'Опыт работы' и последующее
# удаление исходного столбца

resume_base['Опыт работы (Месяц)'] = resume_base['Опыт работы'].apply(get_experience)
resume_base = resume_base.drop('Опыт работы', axis=1)
display(f"Медианный опыт работы (месяцев): {resume_base['Опыт работы (Месяц)'].median()}")
```

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь

# Применяем ранее написанную функцию to_split 
# к столбцу 'Город, переезд, командировки'

resume_base['Город, переезд, командировки'] = \
    resume_base['Город, переезд, командировки']\
    .apply(lambda x: to_split(x, delim=','))

# Формируем столбец 'Город' -  с помощью lambda
# функции берем значение с индексом 0 

resume_base['Город'] = \
    resume_base['Город, переезд, командировки']\
        .apply(lambda x: x[0].strip())

# С помощью lambda функции оптимизируем значения 
# в столбце  'Город' до заданных условием:
# Москва, Санкт-Петербург, город-миллионник, другие

resume_base['Город'] = resume_base['Город']\
    .apply(lambda x: 'Москва' if 'Москва' in x else \
    ('Санкт-Петербург' if 'Санкт-Петербург' in x else\
        ('город-миллионник' if x in ['Новосибирск', \
        'Екатеринбург','Нижний Новгород','Казань',\
        'Челябинск','Омск', 'Самара','Ростов-на-Дону',\
        'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград'] else 'другие')))

# столбец 'Готовность к переезду' - забираем предпоследнее значение с инедксом -2 
#  и переводим в булевы значения True и False в зависимости от готовности к переезду
# Аналогичные преобразования для стобца "Готовность к командировкам"

resume_base['Готовность к переезду'] = \
    resume_base['Город, переезд, командировки']\
        .apply(lambda x: x[1].strip() \
        if 'м.' not in x[1] else x[2].strip())
        
resume_base['Готовность к переезду'] = \
    resume_base['Готовность к переезду']\
        .apply(lambda x: False if 'не г' in x else True)

resume_base['Готовность к командировкам'] = \
    resume_base['Город, переезд, командировки']\
        .apply(lambda x: x[-1].strip() \
        if 'готов' in x[-1] else "не готов(а)")
        
resume_base['Готовность к командировкам'] = \
    resume_base['Готовность к командировкам']\
    .apply(lambda x: False if 'не г' in x else True)

# Удаляем столбец 'Город, переезд, командировки' 
# после получения необходимых данных

resume_base = resume_base.drop('Город, переезд, командировки', axis = 1)

display('Доля соискателей по городам')
display(resume_base['Город'].value_counts(normalize=True))
display("Процент соискателей, готовых и к переезду, и к командировкам:") 
display(resume_base[(resume_base['Готовность к командировкам'] == True)
    & (resume_base['Готовность к переезду'] == True)]\
      .shape[0] / (resume_base.shape[0]) * 100)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

# Создание признаков-мигалок по типу занятости

resume_base['Полная занятость'] = resume_base['Занятость']\
    .apply(lambda x: True if 'полная занятость' in x else False)
resume_base['Частичная занятость'] = resume_base['Занятость']\
    .apply(lambda x: True if 'частичная занятость' in x else False)
resume_base['Проектная работа'] = resume_base['Занятость']\
    .apply(lambda x: True if 'проектная работа' in x else False)
resume_base['Стажировка'] = resume_base['Занятость']\
    .apply(lambda x: True if 'стажировка' in x else False)
resume_base['Волонтерство'] = resume_base['Занятость']\
    .apply(lambda x: True if 'волонтерство' in x else False)

# Создание признаков-мигалок по графику работы

resume_base['Полный день'] = resume_base['График']\
    .apply(lambda x: True if 'полный день' in x else False)
resume_base['Сменный график'] = resume_base['График']\
    .apply(lambda x: True if 'сменный график' in x else False)
resume_base['Гибкий график'] = resume_base['График']\
    .apply(lambda x: True if 'гибкий график' in x else False)
resume_base['Удаленная работа'] = resume_base['График']\
    .apply(lambda x: True if 'удаленная работа' in x else False)
resume_base['Вахтовый метод'] = resume_base['График']\
    .apply(lambda x: True if 'вахтовый метод' in x else False)

# Удаление исходного столбца

resume_base = resume_base.drop(['Занятость', 'График'], axis = 1)

display('Количество соискателей, желающих заниматься \
 проектной работой и волонтерством')
display(resume_base[(resume_base['Проектная работа'] == True) \
    & (resume_base['Волонтерство'] == True)].shape[0])
display('Количество соискателей, рассматривающих \
 вахтовый метод работы и гибкий график')
display(resume_base[(resume_base['Вахтовый метод'] == True) \
    & (resume_base['Гибкий график'] == True)].shape[0])
```

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

# Прочитаем данные из файла формата csv о курсах валют и сохраним
# в датасет exchange_rates

exchange_rates = pd.read_csv('data\ExchangeRates.csv')

# Применим функцию к стобцу 'ЗП' для формирования списка значений
# 

resume_base['ЗП'] = resume_base['ЗП'].apply(to_split)

# Желаемую сумму заработка сохраним в столбец 'ЗП_СУММ', переведя
# в цлеочисленный тип данных

resume_base['ЗП_СУММ'] = resume_base['ЗП'].apply(lambda x: int(x[0]))

# Значение валюты сохраним в столбце 'currency', аналогично
# с названием признака в датафрейме exchange_rates

resume_base['currency'] = resume_base['ЗП'].apply(lambda x: x[1])


def convert_rates(arg):
    """Функция для конвертации наименования валюты
       в формат ISO кодировки

    Args:
        arg (str):текст с наименованием валюты

    Returns:
        arg: Наименование валюты в ISO кодировке
    """
    if arg == 'грн.':
        arg = 'UAH'
    elif arg == 'бел.руб.':
        arg = 'BYN'
    elif arg == 'сум':
        arg = 'UZS'
    elif arg == 'руб.':
        arg = 'RUB'    
    return arg

resume_base['currency'] = resume_base['currency'].apply(convert_rates)

# В исходном датафрейме resume_base создадим признак date с форматом
# datetime, одноименный признак в датафрейме exchange_rates конвертируем
# к такому же формату

resume_base['date'] = pd.to_datetime(resume_base['Обновление резюме'])
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'], dayfirst=True)

# выделим только значение дат, отбросив значение времени

resume_base['date'] = resume_base['date'].dt.date
exchange_rates['date'] = exchange_rates['date'].dt.date

# Создадим датафрейм data_merged путем объединения resume_base
# и exchange_rates методом merge по признакам currency и date

data_merged = resume_base.merge(exchange_rates, on=['currency', 'date'], how='left')

# В столбцах close и proportion заменим значения NaN, полученные по
# строкам с валютой в рублях, на 1

data_merged['close'] =data_merged['close'].fillna(1)
data_merged['proportion'] = data_merged['proportion'].fillna(1)

# Рассчитаем размер заработной платы в рублях и сохраним в признак 'ЗП (руб)'

data_merged['ЗП (руб)'] = data_merged['ЗП_СУММ'] * data_merged['close']\
    / data_merged['proportion']

# Удалим основные и вспомогательные столбцы после преобразования данных

data_merged = data_merged.drop(columns=['ЗП', 'ЗП_СУММ', 'currency', 'date',\
    'per', 'time', 'close', 'vol', 'proportion'])

display(f"Медианная заработная плата составляет: \
    {data_merged['ЗП (руб)'].median()} рублей")
data_merged.info()
```

```python
def outliers_z_score(data, feature, log_scale=False):
    """Функция реализации алгоритма метода z-отклонений

    Args:
        data: датафрейм
        feature: признак, по которому ищутся выбросы
        log_scale: Логарифмирование признака. По умолчанию = false

    Returns:
        outliers: датафрейм со значениями - выбросами
        cleaned: датафрейм, очищенный от выбросов
    """
    
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned


outliers, cleaned = outliers_z_score(drop_data, 'Возраст', log_scale=True)
display(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
display(f'Результирующее число записей: {cleaned.shape[0]}')
display(f"Возраст соискателей, исключенных из массива:{outliers['Возраст'].unique()}")
```
